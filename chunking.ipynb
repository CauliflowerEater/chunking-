```python
# 依赖项
from pathlib import Path
import re
from pathlib import Path
```

```python
raw = Path("dreams.txt").read_text(encoding="utf-8", errors="ignore")
print(raw[:2000])
```

```python
text = raw

# 砍头：从 "版权所有 侵权必究" 标记往上全部不要
start_marker = "版权所有 侵权必究"
start_pos = text.find(start_marker)
if start_pos != -1:
    # 找到标记后，保留标记之后的内容
    text = text[start_pos + len(start_marker):]
else:
    print("警告：未找到起始标记")

# 砍尾：从 "产品经理 | 段冶 封面设计 | 董歆昱" 标记往下全部不要
end_marker = "产品经理 | 段冶 封面设计 | 董歆昱"
end_pos = text.find(end_marker)
if end_pos != -1:
    # 找到标记后，只保留标记之前的内容
    text = text[:end_pos]
else:
    print("警告：未找到结束标记")

# 去除首尾空白
text = text.strip()

print(f"处理后的文本长度: {len(text)} 字符")
print("\n前2000个字符:")
print(text[:2000])
```

```python
#统一换行和空行
text = text.replace('\r\n', '\n').replace('\r', '\n')
lines = text.split('\n')
```

```python
merged_lines = []
buffer = []

for line in lines:
    stripped = line.strip()
    if stripped == "":
        # 段落结束
        if buffer:
            merged_lines.append(" ".join(buffer))
            buffer = []
        merged_lines.append("")  # 保留空行作为段落分隔符
    else:
        buffer.append(stripped)

# 文件末尾收尾
if buffer:
    merged_lines.append(" ".join(buffer))

text = "\n".join(merged_lines)
```

```python
print(text[:2000])
```

```python

# 清掉行首尾多余空格
text = "\n".join(line.rstrip() for line in text.split("\n"))

# 将 3 行及以上连续空行压缩成 2 行
text = re.sub(r'\n{3,}', '\n\n', text)
```

```python
#保存清洗结束之后的文本

Path("dreams_clean.txt").write_text(text, encoding="utf-8")
```
