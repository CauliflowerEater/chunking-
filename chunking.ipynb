```python
# 依赖项
from pathlib import Path
import re
from pathlib import Path
```

```python
raw = Path("dreams.txt").read_text(encoding="utf-8", errors="ignore")
print(raw[:2000])
```

```python
text = raw

# 砍头：从 "版权所有 侵权必究" 标记往上全部不要
start_marker = "版权所有 侵权必究"
start_pos = text.find(start_marker)
if start_pos != -1:
    # 找到标记后，保留标记之后的内容
    text = text[start_pos + len(start_marker):]
else:
    print("警告：未找到起始标记")

# 砍尾：从 "产品经理 | 段冶 封面设计 | 董歆昱" 标记往下全部不要
end_marker = "产品经理 | 段冶 封面设计 | 董歆昱"
end_pos = text.find(end_marker)
if end_pos != -1:
    # 找到标记后，只保留标记之前的内容
    text = text[:end_pos]
else:
    print("警告：未找到结束标记")

# 去除首尾空白
text = text.strip()

print(f"处理后的文本长度: {len(text)} 字符")
print("\n前2000个字符:")
print(text[:2000])
```

```python
#统一换行和空行
text = text.replace('\r\n', '\n').replace('\r', '\n')
lines = text.split('\n')
```

```python
merged_lines = []
buffer = []

for line in lines:
    stripped = line.strip()
    if stripped == "":
        # 段落结束
        if buffer:
            merged_lines.append(" ".join(buffer))
            buffer = []
        merged_lines.append("")  # 保留空行作为段落分隔符
    else:
        buffer.append(stripped)

# 文件末尾收尾
if buffer:
    merged_lines.append(" ".join(buffer))

text = "\n".join(merged_lines)
```

```python
import json

# 读取清理后的文本
text = Path("dreams_clean.txt").read_text(encoding="utf-8")
lines = text.split('\n')

# 存储段落的列表
chapters = []
current_chapter_id = 0
current_chapter_name = None
current_chapter_text = []
skip_next_empty = False  # 标志：是否跳过下一个空行（章节名后的结构性空行）
i = 0

while i < len(lines):
    line = lines[i].strip()
    
    # 检查是否是段落号：Chapter XX 或 初版序言
    if line.startswith("Chapter ") or line == "初版序言":
        # 如果之前有正在收集的段落，先保存它
        if current_chapter_name is not None:
            chapter_text = '\n'.join(current_chapter_text).strip()
            if chapter_text:  # 只保存非空段落
                chapters.append({
                    "chapter-ID": current_chapter_id,
                    "chapter-name": current_chapter_name,
                    "chapter-text": chapter_text
                })
                current_chapter_id += 1
            current_chapter_text = []
        
        # 记录段落号
        chapter_number = line
        
        # 跳过当前行，检查下一行是否为空行
        i += 1
        if i < len(lines) and lines[i].strip() == "":
            # 空行后，检查下一行是否是段落名（不是Chapter开头）
            i += 1
            if i < len(lines):
                next_line = lines[i].strip()
                # 如果下一行不是新的段落号且当前段落不是序言，则它是段落名
                if not (next_line.startswith("Chapter ") or chapter_number == "初版序言"):
                    current_chapter_name = next_line
                    i += 1  # 移动到段落名之后
                    # 设置标志：下一行如果是空行，应该跳过（这是章节名后的结构性空行）
                    skip_next_empty = True
                else:
                    # 如果下一行也是段落号，或者当前段落是序言，说明这个段落没有段落名
                    current_chapter_name = chapter_number
                    # i 回退到空行，下一轮循环会处理这个空行
                    i -= 1
                    skip_next_empty = True  # 序言后的空行也应该跳过
            else:
                current_chapter_name = chapter_number
        else:
            # 如果没有空行，段落号就是段落名
            current_chapter_name = chapter_number
    else:
        # 收集段落内容
        if current_chapter_name is not None:
            # 如果设置了跳过空行的标志，且当前行是空行，则跳过
            if skip_next_empty and line == "":
                skip_next_empty = False  # 重置标志
            else:
                # 正常收集内容（包括段落内容中的空行）
                current_chapter_text.append(lines[i])
                skip_next_empty = False  # 重置标志
        i += 1

# 保存最后一个段落
if current_chapter_name is not None:
    chapter_text = '\n'.join(current_chapter_text).strip()
    if chapter_text:  # 只保存非空段落
        chapters.append({
            "chapter-ID": current_chapter_id,
            "chapter-name": current_chapter_name,
            "chapter-text": chapter_text
        })

# 转换为JSON格式
chapters_json = json.dumps(chapters, ensure_ascii=False, indent=2)

# 保存到文件
Path("chapters.json").write_text(chapters_json, encoding="utf-8")

print(f"共识别到 {len(chapters)} 个章节")
print("\n前3个章节的信息:")
for i, chapter in enumerate(chapters[:3]):
    print(f"\n章节ID: {chapter['chapter-ID']}")
    print(f"章节名: {chapter['chapter-name']}")
    print(f"章节文本长度: {len(chapter['chapter-text'])} 字符")
    print(f"章节文本预览: {chapter['chapter-text'][:100]}...")
```

```python
# 读取现有的 chapters.json
import json

with open("chapters.json", "r", encoding="utf-8") as f:
    chapters = json.load(f)

# 对每个章节的段落文本进行拆分
chapters_with_paragraphs = []

for chapter in chapters:
    chapter_text = chapter["chapter-text"]
    
    # 按空行分割段落（保留空行作为分隔符）
    paragraphs_list = []
    current_paragraph = []
    
    lines = chapter_text.split('\n')
    
    for line in lines:
        if line.strip() == "":
            # 遇到空行，如果当前段落有内容，则保存
            if current_paragraph:
                paragraph_text = ' '.join(current_paragraph).strip()
                if paragraph_text:  # 只保存非空段落
                    paragraphs_list.append(paragraph_text)
                current_paragraph = []
        else:
            current_paragraph.append(line)
    
    # 保存最后一个段落
    if current_paragraph:
        paragraph_text = '\n'.join(current_paragraph).strip()
        if paragraph_text:
            paragraphs_list.append(paragraph_text)
    
    # 构建新的章节结构，包含 paragraphs 数组
    chapter_with_paragraphs = {
        "chapter-ID": chapter["chapter-ID"],
        "chapter-name": chapter["chapter-name"],
        "paragraphs": [
            {
                "paragraph-id": idx,  # 从1开始
                "text": para_text
            }
            for idx, para_text in enumerate(paragraphs_list)
        ]
    }
    
    chapters_with_paragraphs.append(chapter_with_paragraphs)

# 保存到新文件
chapters_with_paragraphs_json = json.dumps(chapters_with_paragraphs, ensure_ascii=False, indent=2)
Path("chapters_with_paragraphs.json").write_text(chapters_with_paragraphs_json, encoding="utf-8")

print(f"处理完成！共 {len(chapters_with_paragraphs)} 个章节")
print("\n前2个章节的段落统计:")
for chapter in chapters_with_paragraphs[:2]:
    print(f"\n章节 {chapter['chapter-ID']}: {chapter['chapter-name']}")
    print(f"  包含 {len(chapter['paragraphs'])} 个小段落")
    if chapter['paragraphs']:
        print(f"  第一个段落预览: {chapter['paragraphs'][0]['text'][:100]}...")
```

```python
with open("chapters_with_paragraphs.json", "r", encoding="utf-8") as f:
    chapters = json.load(f)

MAX_CHARS = 1000   # 一个 chunk 目标最大长度，你可以根据实际调
MIN_CHARS = 300    # 用来处理最后太短的 chunk，可选

chunks = []
global_chunk_index = 0  # 全书范围的 chunk 序号

for chapter in chapters:
    chapter_id = chapter["chapter-ID"]
    chapter_title = chapter["chapter-name"]
    paras = chapter["paragraphs"]

    current_paras = []
    current_len = 0
    chunk_index_in_chapter = 0

    for para in paras:
        text = para["text"].strip()
        plen = len(text)

        # 如果单个段落就已经比 MAX_CHARS 还多，那它就单独成一个 chunk
        if plen >= MAX_CHARS:
            # 先把之前累积的 flush 掉
            if current_paras:
                chunks.append({
                    "chunk_index": global_chunk_index,
                    "chunk_index_in_chapter": chunk_index_in_chapter,
                    "chapter-ID": chapter_id,
                    "chapter-name": chapter_title,
                    "para_start": current_paras[0]["paragraph-id"],
                    "para_end": current_paras[-1]["paragraph-id"],
                    "text": "\n\n".join(p["text"] for p in current_paras),
                })
                global_chunk_index += 1
                chunk_index_in_chapter += 1
                current_paras = []
                current_len = 0

            # 这个长段落自己一个 chunk
            chunks.append({
                "chunk_index": global_chunk_index,
                "chunk_index_in_chapter": chunk_index_in_chapter,
                "chapter-ID": chapter_id,
                "chapter-name": chapter_title,
                "para_start": para["paragraph-id"],
                "para_end": para["paragraph-id"],
                "text": text,
            })
            global_chunk_index += 1
            chunk_index_in_chapter += 1
            continue

        # 正常情况：尝试把这个段落塞进当前 chunk
        if current_len + plen <= MAX_CHARS:
            current_paras.append(para)
            current_len += plen
        else:
            # 当前 chunk 到上限了，先 flush
            # 保存上一个chunk的最后2个段落用于overlapping
            last_two_paras = []
            if current_paras:
                # 保存最后2个段落（如果存在）
                last_two_paras = current_paras[-2:] if len(current_paras) >= 2 else current_paras[:]
                
                chunks.append({
                    "chunk_index": global_chunk_index,
                    "chunk_index_in_chapter": chunk_index_in_chapter,
                    "chapter-ID": chapter_id,
                    "chapter-name": chapter_title,
                    "para_start": current_paras[0]["paragraph-id"],
                    "para_end": current_paras[-1]["paragraph-id"],
                    "text": "\n\n".join(p["text"] for p in current_paras),
                })
                global_chunk_index += 1
                chunk_index_in_chapter += 1

            # 用当前段落开新 chunk，并包含上一个chunk的最后2个段落（overlapping）
            current_paras = last_two_paras + [para]
            # 计算新chunk的长度：上一个chunk最后2个段落 + 当前段落
            current_len = sum(len(p["text"]) for p in last_two_paras) + plen

    # 章节结束，收尾
    if current_paras:
        chunks.append({
            "chunk_index": global_chunk_index,
            "chunk_index_in_chapter": chunk_index_in_chapter,
            "chapter-ID": chapter_id,
            "chapter-name": chapter_title,
            "para_start": current_paras[0]["paragraph-id"],
            "para_end": current_paras[-1]["paragraph-id"],
            "text": "\n\n".join(p["text"] for p in current_paras),
        })
        global_chunk_index += 1

chunks_json = json.dumps(chunks, ensure_ascii=False, indent=2)
Path("chunks.json").write_text(chunks_json, encoding="utf-8")

print(f"共生成 {len(chunks)} 个 chunk")
print("\n前3个 chunk 的预览:")
for i, chunk in enumerate(chunks[:3]):
    print(f"\nChunk {chunk['chunk_index']}:")
    print(f"\nChunk_index_in_chapter: {chunk['chunk_index_in_chapter']}")
    print(f"\nChapter_ID: {chunk['chapter-ID']}")
    print(f"\nChapter_name: {chunk['chapter-name']}")
    print(f"\nPara_start: {chunk['para_start']}")
    print(f"\nPara_end: {chunk['para_end']}")
    print(f"\nText: {chunk['text']}")
```
